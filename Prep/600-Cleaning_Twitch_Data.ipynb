```python
import pandas as pd
import csv
import os
```

<b> In this notebook, we are going to combine all the 4 years of Twitch Data and drop all the unnecessary columns.</b>

```python
t_2016 = pd.read_csv("C:\\Users\\gugla\\Desktop\\Twitch_Data\\Raw_Data\\twitch_2016.csv")
```

```python
# When removing all the columns together it was giving an error, so had to remove them separately.
t_2016 = t_2016.drop(['Unnamed: 0'],axis = 1)
t_2016 = t_2016.drop(['Unnamed: 0.1'],axis = 1)
t_2016 = t_2016.drop(['Unnamed: 1'],axis = 1) 
t_2016 = t_2016.drop(['Unnamed: 11'],axis = 1)
```

```python
t_2016
```

```python
t_2017= pd.read_csv("C:\\Users\\gugla\\Desktop\\Twitch_Data\\Raw_Data\\twitch_2017.csv")
```

```python
t_2017 = t_2017.drop(['Unnamed: 0'],axis = 1)
t_2017 = t_2017.drop(['Unnamed: 0.1'],axis = 1)
t_2017 = t_2017.drop(['Unnamed: 1'],axis = 1) 
t_2017 = t_2017.drop(['Unnamed: 11'],axis = 1)
```

```python
t_2017
```

```python
t_2018 = pd.read_csv("C:\\Users\\gugla\\Desktop\\Twitch_Data\\Raw_Data\\twitch_2018.csv")
```

```python
t_2018 = t_2018.drop(['Unnamed: 0'],axis = 1)
t_2018 = t_2018.drop(['Unnamed: 0.1'],axis = 1)
t_2018 = t_2018.drop(['Unnamed: 1'],axis = 1) 
t_2018 = t_2018.drop(['Unnamed: 11'],axis = 1)
```

```python
t_2018
```

```python
t_2019 = pd.read_csv("C:\\Users\\gugla\\Desktop\\Twitch_Data\\Raw_Data\\twitch_2019.csv")
```

```python
t_2019 = t_2019.drop(['Unnamed: 0'],axis = 1)
t_2019 = t_2019.drop(['Unnamed: 0.1'],axis = 1)
t_2019 = t_2019.drop(['Unnamed: 1'],axis = 1) 
t_2019 = t_2019.drop(['Unnamed: 11'],axis = 1)
```

```python
t_2019
```

```python
t_16_17 = t_2016.merge(t_2017,how='outer')
```

```python
t_16_17
```

```python
t_16_17_18 = t_16_17.merge(t_2018,how='outer')
```

```python
t_16_17_18
```

```python
twitch_final = t_16_17_18.merge(t_2019,how='outer')
```

```python
twitch_final
```

<b> Renaming the columns</b>

```python
twitch_final.rename(columns = {
    'Watch time': 'Watch_time',
    'Stream time': 'Stream_time',
    'Peak viewers': 'Peak_viewers',
    'Peak channels': 'Peak_channels',
    'Average viewers': 'Average_viewers',
    'Average channels': 'Average_channels',
    'Average viewer ratio': 'Average_viewer_ratio'
},inplace = True)
```

<b>Grouping the dataframe by Game and adding all the columns and sorting the dataframe by watch time so the most watched game comes on the top.</b>

```python
twitch_final = twitch_final.groupby('Game').sum().sort_values(by='Watch_time',ascending=False).reset_index()
```

```python
twitch_final
```

```python
twitch_final.to_csv("twitch_final.csv")
```
