```python
vg_df= pd.read_csv("../data/prep/100-Collecting_VGChartz.csv", low_memory = False)
meta_df= pd.read_csv("../data/raw/Original_MetaCritic.csv", low_memory = False)
```

## **Joining VGChartz and MetaCritic datasets**
I will be joining the datasets by metacritic url. In our first notebook I created a metacritic url for each game that we collected from VGChartz. I added these urls to the VGChartz dataframe in a column called 'meta_url'. Then when scraping the game data from MetaCritic I also added the url used for each game in a column called 'meta_full_url'. I am going the join the two dataframe by these columns.

```python
meta_vg=vg_df.merge(meta_df, left_on='meta_url', right_on='meta_full_url')
meta_vg=meta_vg.drop(['meta_url'], axis = 1)
```
