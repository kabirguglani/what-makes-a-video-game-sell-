```python
import pandas as pd
import csv
import os
```

<b> t = twitch Dataset, 
    v = VG/Meta Dataset</b>

```python
#t = pd.read_csv("Twitch_final.csv")
t = pd.read_csv("../data/raw/final_twitch.csv")
```

```python
t.head(5)
```

```python
t = t.drop(['Unnamed: 0'],axis = 1)
```

```python
v = pd.read_csv("../data/prep/400.csv")
```

```python
v.head(5)
```

<b> Renaming the names of Game column on both the datasets and also converting the game names to lowercase, so we don't have any problem while merging both the datasets.</b>

```python
v =v.rename(columns={'name': 'game'})
t=t.rename(columns={'Game': 'game'})   
```

```python
t['game'] = t['game'].str.lower()
v['game'] = v['game'].str.lower()
```

```python
# Using strip function to remove the unnecessary spaces
v.game = v.game.str.strip()
```

```python
v.head(5)
```

```python
combined_df = pd.merge(t,v,on='game')
```

```python
combined_df
```

```python
combined_df.to_csv("../data/prep/700-Combined_VG_Twitch.csv")
```
